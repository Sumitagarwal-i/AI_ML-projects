```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix


iris = load_iris()

print(iris.keys())


# For easy visualization
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['target'] = iris.target
df.head()

X = df.drop('target',axis = 1)
Y = df['target']

# print(X)
# print(Y)

X_train,X_test,Y_train,Y_test = train_test_split(
    X,Y,test_size = 0.2,random_state=42
)

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train,Y_train)

Y_pred = model.predict(X_test)

print("Confusion Matrix")
print(confusion_matrix(Y_test,Y_pred))

print("Classification report")
print(classification_report(Y_test,Y_pred))







```

# KNN (K- Nearest Neighbours) :

A KNN model is a simple , non-parametric, supervised machine learning algorithm used for both classification and regression tasks.

It works by finding the 'K' closest data points(neighbours) to a new, unseen data point and making predictions based on these neighbours.

K -> It is a user defined integer

For classification , KNN assigns the most common class among these neighbours to the new data point.

For regression , it predicts the average value of neighbours.

Summary : A KNN models predicts the label or value of a new data point by looking at the labels or values of its k nearest neighbours in the training data , relying on the principle that similar points are likely to have similar outcomes.

```python
#Visualizing

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix

iris = load_iris()

df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['target'] = iris.target
df.head()

X = df.drop('target',axis = 1)
Y = df['target']

# * Visualizing the models predictions ->  For only 2 features
X_vis = df[['sepal length (cm)','sepal width (cm)']]
Y_vis = df['target']


X_train,X_test,Y_train,Y_test = train_test_split(
    X_vis,Y_vis,test_size = 0.2,random_state = 42
)

model = KNeighborsClassifier(n_neighbors = 3)
model.fit(X_train,Y_train)

Y_pred = model.predict(X_test)


plt.figure(figsize = (8,6))

plt.scatter(X_test['sepal length (cm)'],X_test['sepal width (cm)'], c = Y_pred , cmap='viridis' , s=100,edgecolors = 'k')

plt.title('KNN Predictions (k=3) in 2D Feature Space')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True)
plt.show()

```
